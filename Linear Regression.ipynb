```python
 # Dependencies
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
data_path = "data.csv"
```

```python
cancer_df = pd.read_csv(data_path, encoding="ISO-8859-1")
```

```python
cancer_df.head()
```

```python
null_columns=cancer_df.columns[cancer_df.isnull().any()]
cancer_df[null_columns].isnull().sum()
```

```python
cancer_df.head()
```

```python
df = cancer_df.drop(['Unnamed: 32'], axis=1)
```

```python
df['diagnosis'] = df['diagnosis'].map({'M':1,'B':0})
df.head()
```

```python
data_final = df.drop("id", axis=1)
feature_names = data_final.columns
data_final.head()
```

```python
df_final = data_final[['diagnosis','area_worst','radius_worst', ]]
df_final.head()
```

```python
X = df_final.drop("diagnosis", axis=1)
y = df_final["diagnosis"]

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

model.summary()
```
